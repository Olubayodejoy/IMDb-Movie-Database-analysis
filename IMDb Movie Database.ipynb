```python
import pandas as pd
import seaborn as sns
IMDb_Movie = pd.read_csv('C:\\Users\\HP\\Documents\\IMDB Top 250 Movies.csv')
IMDb_Movie
```

```python
filtered9_df = IMDb_Movie[IMDb_Movie['year'] > 2000]
filtered9_df
```

```python
def rating_categories(rate):
    if rate >= 9.5:
        return 'Excellent'
    elif rate >= 9:
        return 'Good'
    elif rate >= 8.5:
        return 'Average'
    else:
        return 'Okay'

IMDb_Movie['rating categories'] = IMDb_Movie['rating'].apply(rating_categories)

IMDb_Movie
```
